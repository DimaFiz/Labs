*–ü—Ä–µ–¥–µ–ª –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º `xlim` –ª–∏–±–æ `ylim`.*

*–î–ª—è —É–¥–æ–±—Å—Ç–≤–∞ –ø–æ–ª—å–∑—É–π—Ç–µ—Å—å —Ü–∏–∫–ª–æ–º `for`, —á—Ç–æ–±—ã —Å–æ–∑–¥–∞—Ç—å `tuple`, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ `tuple`, —Å–æ–¥–µ—Ä–∂–∞—â–∏—Ö –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–∏ –∏ `tuple` –∏–∑ –∏–º–µ–Ω–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π, –ª–µ–≤–æ–≥–æ –∏ –ø—Ä–∞–≤–æ–≥–æ –∫–æ–Ω—Ü–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞.*

*P.S. $\cot x$ —ç—Ç–æ –∫–æ—Ç–∞–Ω–≥–µ–Ω—Å $x$*

```python
# –í —ç—Ç–æ–º –∑–∞–¥–∞–Ω–∏–∏ –≤—ã –º–æ–∂–µ—Ç–µ –≤—ã–±—Ä–∞—Ç—å –ª—é–±–æ–π —Å—Ç–∏–ª—å –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏—è —Å–≤–æ–µ–≥–æ –∫–æ–¥–∞

def my_cot(x):
    return sp.cos(2 * x + 3)/sp.sin(2 * x + 3)





def my_plot(start, end):
    p = plot(show=False)
    x = Symbol("x")
    p.ylim = (-5, 5)

    a = (2 * start + 3)//(np.pi) +1 if (2 * start + 3) - (np.pi) *( (2 * start + 3)//(np.pi)) > 1e-1 else  2 * (2 * start + 3)//(np.pi/2)
    b = (2 * end + 3)//(np.pi)
    
    last = start

    if  a * (np.pi/2)- (start + 3) > 1e-1:
        p.append(plot(my_cot(x), (x, a * np.pi/2 - 1.5- 0.01), line_color="blue", show=False)[0])

    last = a 

    for i in range(int(a)+ 1, int(b)+1):
        p.append(plot(my_cot(x), (x, last * (np.pi/2) + 0.01  - 1.5, i *(np.pi/2) - 1.5 - 0.01), line_color="blue", show=False)[0])
        last = i

    if (end + 3) - b * (np.pi/2) > 1e-2:
        p.append(plot(my_cot(x), (x, b *(np.pi/2) +  - 1.5 +0.01, end), line_color="blue", show=False)[0])

        


    
    p.show()

my_plot(0, 7 * np.pi)
    

        


# ·É¶„Çù‚ó°‚ïπ)„Éé‚ô° –í–∞—à –∫–æ–¥ –∑–¥–µ—Å—å

```

–í—ã–∏—á–∏—Å–ª–∏–º –ø—Ä–µ–¥–µ–ª —á–µ—Ä–µ–∑ –∞
$$
\lim_{x \to 0} \frac{\sin x - x \cos(ax)}{x^3} = \lim_{x \to 0} \frac{x -\frac{x^3}{6} + o(x^3)- x (1 - \frac{(ax)^2}{2} + o(x^2))}{x^3} = \lim_{x \to 0} \frac{ -\frac{x^3}{6} + \frac{a^2 x^3}{2} + o(x^3)}{x^3} = \frac{a^2}{2} - \frac{1}{6}.

$$
\frac{a^2}{2} - \frac{1}{6} = 2
\\
a = \pm \sqrt{\frac{13}{3}}
$$

```python
alpha_8 = 0 # –í–º–µ—Å—Ç–æ –Ω—É–ª—è –ø—Ä–∏—Å–≤–æ–π—Ç–µ –æ—Ç–≤–µ—Ç

x = sp.symbols("x")
a = sp.symbols("a", real=True)

expression = (sp.sin(x) - x * (sp.cos(a * x))) / (x ** 3)

limit_res = sp.limit(expression, x, 0)



equation = sp.Eq(limit_res, 2)

ans = sp.solve(equation, a)

alpha_8 = ans
display(*alpha_8)
```

**<font color=#FF5EDD>–ó–∞–¥–∞—á–∞ 10**</font> (3 –±–∞–ª–ª–∞)

–ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, —É –í–∞—Å –µ—Å—Ç—å –Ω–∞—á–∞–ª—å–Ω–∞—è —Å—É–º–º–∞ –¥–µ–Ω–µ–≥ `initial`, –∏ –í—ã —Ö–æ—Ç–∏—Ç–µ –∏–Ω–≤–µ—Å—Ç–∏—Ä–æ–≤–∞—Ç—å –µ–µ –ø–æ–¥ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≥–æ–¥–æ–≤–æ–π –ø—Ä–æ—Ü–µ–Ω—Ç–Ω—ã–π –¥–æ—Ö–æ–¥ `extra`. –í–∞–º –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏, —Å–∫–æ–ª—å–∫–æ –≤—Ä–µ–º–µ–Ω–∏ `time` –ø–æ—Ç—Ä–µ–±—É–µ—Ç—Å—è, —á—Ç–æ–±—ã –í–∞—à–∞ –∏–Ω–≤–µ—Å—Ç–∏—Ü–∏—è –¥–æ—Å—Ç–∏–≥–ª–∞ –∏–ª–∏ –ø—Ä–µ–≤—ã—Å–∏–ª–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—É—é —Ü–µ–ª–µ–≤—É—é —Å—É–º–º—É `final`.

 –£—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–ª—è —ç—Ç–æ–π –∑–∞–¥–∞—á–∏ –≤—ã–≥–ª—è–¥–∏—Ç —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

$$ initial √ó \left(1 + \frac{extra}{100}\right)^{time} = final $$

–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `check_time(initial: float, extra: float, final: float) -> int` —á—Ç–æ–±—ã –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å —Å–∫–æ–ª—å–∫–æ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ–Ω–∞–¥–æ–±–∏—Ç—Å—è —á—Ç–æ–±—ã –¥–æ—Å—Ç–∏—á—å –∂–µ–ª–∞–µ–º–æ–π —Ü–µ–ª–∏.

```python
def check_time(initial, extra, final) -> int:
    time = 0 # –í–º–µ—Å—Ç–æ –Ω—É–ª—è –∑–∞–ø–∏—à–∏—Ç–µ –∏—Å–∫–æ–º–æ–µ –∫–æ–ª–∏—á–∏—Å—Ç–≤–æ –≤—Ä–µ–º–µ–Ω–∏

    x = sp.symbols("x")

    expression = initial * ((1 + extra / 100) ** (x))
    equation = sp.Eq(expression, final)
    res = sp.solve(equation, x)
    if res:
        time = np.ceil(res[0])
    else:
        time = None
    return time

```
