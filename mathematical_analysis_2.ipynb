```python

x = Symbol('x')
def closest_point(f_1, x_0, y_0):
  result = () #type(result) = tuple = (x-координата искомой точки, у-координата искомой точки). В эту переменную запиши координаты искомой точки.
  
  dist = (x-x_0)**2 + (f_1 - y_0)**2
  ext = (solve(diff(dist, x), x))
  
  min_ = sp.oo
  x_best = y_best = 0
  for x_ in ext:
    if not x_.is_real:
      continue
    d = ((x_ -x_0)**2 + (f_1.subs(x, x_) - y_0)**2).evalf()
    if d < min_:
      min_ = dist
      x_best = x_
      y_best = f_1.subs(x, x_)


  result = (x_best, y_best)
  return result


```

Гонщик Гоша движется по извилистой трассе, заданной параметрически уравнениями:
$$
\left\{

y = \sin(4t) -0.3\cos(3t)

$$
где $ t $ — параметр, изменяющийся на интервале от $ t = 0 $ до $ t = 2 \pi $.

Помогите Гоше:

1. Определить все точки $t_{\text{max}}$ и $t_{\text{min}}$, в которых перегрузка $a(t)$ достигает своего максимального и минимального значения (перегрузка пропорциональна модулю длины вектора ускорения (это и есть вектор из вторых производных)).
2. Построить функцию $f(t)$, которая по времени $t$ выдает величину перегрузки в доле от минимальной до максимальной
(то есть
 $f(t) = \frac{a(t) - a(t_{\text{min}})}{a(t_{\text{max}}) - a(t_{\text{min}})}$).
3. Построить график трассы с градиентной раскраской от зеленого цвета(минимальной перегрузки: $f(t) = 0$) до красного цвета (максимальной перегрузки: $f(t) = 1$). Воспользуйтесь `cmap = plt.cm.RdYlGn_r`. Отметьте на трассе все точки с минимальной и  максимальной перегрузками.

```python
t = Symbol('t')

t_min_4 = []  #type(t_min_4) = list. В эту переменную вместо [] запишите все точки с минимальной перегрузкой
t_max_4 = []  #type(t_max_4) = list. В эту переменную вместо [] запишите все точки с максимальной перегрузкой




f_4 = t  #type(f_4) = sympy.core.mul.Mul. В эту переменную вместо t запишите функцию f(t)

# ღゝ◡╹)ノ♡ Ваш код здесь

# Через Solve не получается оно не может решить уравнение
# поэтому дискретизируем точки


# Параметрические уравнения
x_sym = Cos(4*t) + 0.6*Sin(2*t)
y_sym = Sin(4*t) - 0.3*Cos(3*t)

# Первые производные (скорость) - для проверки, но для перегрузки нужны вторые
vx_sym = diff(x_sym, t)
vy_sym = diff(y_sym, t)

# Вторые производные (ускорение)
ax_sym = diff(vx_sym, t)
ay_sym = diff(vy_sym, t)

# Модуль ускорения (перегрузка)
a = Sqrt(ax_sym**2 + ay_sym**2)

# Создаем быструю функцию для вычислений
diff_acc_sq = diff(a, t)



t_vals = np.linspace(0, 2 * np.pi, 3000)
acc_vals = np.array([float(a.subs(t, i).evalf()) for i in t_vals])

# Находим глобальные минимум и максимум на сетке
glob_min_approx = np.min(acc_vals)
glob_max_approx = np.max(acc_vals)

# Ищем индексы всех локальных экстремумов, которые близки к глобальным
# Используем допуск (tolerance), так как сетка дискретна
tolerance = 0.05 

t_min_guesses = []
t_max_guesses = []

for i in range(1, len(acc_vals) - 1):
    # Локальный минимум
    if acc_vals[i] < acc_vals[i-1] and acc_vals[i] < acc_vals[i+1]:
        if abs(acc_vals[i] - glob_min_approx) < tolerance:
            t_min_guesses.append(t_vals[i])
            
    # Локальный максимум
    if acc_vals[i] > acc_vals[i-1] and acc_vals[i] > acc_vals[i+1]:
        if abs(acc_vals[i] - glob_max_approx) < tolerance:
            t_max_guesses.append(t_vals[i])

t_min_4 = t_min_guesses
t_max_4 = t_max_guesses

t_min_4.sort()
t_max_4.sort()

min_val_exact = a.subs(t, t_min_4[0])
max_val_exact = a.subs(t, t_max_4[0])


f_4 = (a - min_val_exact) / (max_val_exact - min_val_exact)


x_plot = np.array([float(x_sym.subs(t, i).evalf()) for i in t_vals])
y_plot = np.array([float(y_sym.subs(t, i).evalf()) for i in t_vals])
colors =  np.array([float(f_4.subs(t, i).evalf()) for i in t_vals])





plt.figure(figsize=(10, 8))


plt.scatter(x_plot, y_plot, c=colors, cmap=plt.cm.RdYlGn_r, s=5)

# Отмечаем точки минимума
x_min = [x_sym.subs(t, val) for val in t_min_4]
y_min = [y_sym.subs(t, val) for val in t_min_4]
plt.scatter(x_min, y_min, color='green', s=100, edgecolors='black', label='Минимальная перегрузка', zorder=10)

# Отмечаем точки максимума
x_max = [x_sym.subs(t, val) for val in t_max_4]
y_max = [y_sym.subs(t, val) for val in t_max_4]
plt.scatter(x_max, y_max, color='red', s=100, edgecolors='black', label='Максимальная перегрузка', zorder=10)

plt.colorbar(label='Нормированная перегрузка f(t)')
plt.title('Трасса Гоши(Георгия)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axis('equal')
plt.grid(True, alpha=0.3)


print(f"Точки минимума (t): ", *t_min_4)
print(f"Точки максимума (t): ", *t_max_4)
print(f"Функция f(t):", end=" ")
display(f_4)

plt.show()
```
